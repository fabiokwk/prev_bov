# Prever fechamento do bovespa


### A finalidade desse código é preparar um dataset para utilizar como base de um modelo que prevê o fechamento diário  do bovespa.

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np
from bcb import sgs
from datetime import datetime
```

```python
# data de início e fim de coleta dos dados
start = '2003-12-01'
end = '2023-10-01'
# tickers que serão utilizados no modelo
tickers = ["^BVSP", "^GSPC", "BRL=X", "^N225", "^NYA", "^IXIC", "^HSI", "^FTSE"]
# df com fechamento do bovespa e abertura dos demais
prev_bov = pd.DataFrame()
# iteração para salvar os valores conforme estipulado 
for ticker in tickers:
    data = yf.download(ticker, start=start, end=end)
    if ticker != "^BVSP":
        prev_bov[ticker]=data['Open']
    else:
        prev_bov[ticker]=data['Close']
    
```

```python
prev_bov.tail()
```

```python
# excluindo linhas que contenham pelo menos 1 valor nulo
prev_bov = prev_bov.dropna(how='any')
```

```python
# renomeando as variáveis 
prev_bov.rename(columns={'^BVSP': 'bovespa',
                        '^GSPC': 'sp500',
                        'BRL=X': 'dolar',
                        '^N225': 'nikkei',
                        '^NYA': 'newyork',
                        '^IXIC': 'nasdaq',
                        '^HSI': 'xangai',
                        '^FTSE': 'londres',}, inplace=True)
```

```python
prev_bov = prev_bov.round(2)
prev_bov
```

```python
# exportar resultado para um CSV, que será analisado com R
prev_bov.to_csv('prev_bov.csv', index=True)
```
